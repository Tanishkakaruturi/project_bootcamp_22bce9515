```python
top_rated_shows = newdf.sort_values(by='imdb_score', ascending=False).head(10)
plt.figure(figsize=(6, 6))
plt.bar(top_rated_shows['title'], top_rated_shows['imdb_score'], color='pink')
plt.title('Top 10 Netflix Shows Based on imdb_score')
plt.xlabel('title')
plt.ylabel('imdb_score')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
yearly_distribution = df['runtime'].value_counts().sort_index()
plt.plot(yearly_distribution.index, yearly_distribution.values, marker='o', color='skyblue')
plt.title('Graph plotted based on runtime of nextflix shows')
plt.xlabel('runtime')
plt.ylabel('Number of Shows')
plt.grid(True)
plt.show()
```

```python
age_cert_avg_scores = df.groupby('age_certification')['imdb_score'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
age_cert_avg_scores.plot(kind='bar', color='skyblue')
plt.title('IMDb Scores by Age Certification')
plt.xlabel('age_certification')
plt.ylabel('imdb_score')
plt.xticks(rotation=0)  # Keep the x-axis labels horizontal
plt.show()
```

```python
movie_data= df.head(100)
plt.figure(figsize=(6, 6))
plt.scatter(movie_data['runtime'], movie_data['imdb_score'], color='purple', alpha=0.7)
plt.title('Scatter Plot of Movie Runtime vs IMDb Score')
plt.xlabel('Movie Runtime (minutes)')
plt.ylabel('IMDb Score')
plt.grid(True)
plt.show()
```

```python
netflix_data= df.head(100)
plt.figure(figsize=(12, 6))
certification_counts = netflix_data.groupby(['release_year', 'age_certification']).size().unstack(fill_value=0)
certification_percentages = certification_counts.div(certification_counts.sum(axis=1), axis=0) * 100
certification_percentages.plot(kind='bar', stacked=True, colormap='viridis', width=0.8)
plt.title('Distribution of Age Certifications Over Time')
plt.xlabel('Release Year')
plt.ylabel('Percentage of Shows')
plt.legend(title='Age Certification', bbox_to_anchor=(1, 1))
plt.show()
```
